```python
def check_min_samples(class_objs_list, classes):
    
    meta_data = pd.DataFrame(columns=["N° objs","N° samples", "N° min-max samples (g)", "N° min-max samples (r)"])
    for idx, class_objs in enumerate(class_objs_list):
        min_samples_g = True
        min_samples_r = True
        n_g = []
        n_r = []
        for idx, obj_id in enumerate(class_objs.objs_df.index.unique()):
            obj = class_objs.objs_df.loc[obj_id]
            obj_g = obj[obj.fid == 1]
            n_g.append(obj_g.shape[0])
            obj_r = obj[obj.fid == 2]
            n_r.append(obj_r.shape[0])
            if obj_g.shape[0] < 60:
                min_samples_g = False
                print(f"{obj_id} with less samples for g-band")
            if obj_r.shape[0] < 60:
                min_samples_r = false
                print(f"{obj_id} with less samples for r-band")
        meta_data = meta_data.append({"N° objs":class_objs.n_objs,
                                      "N° samples":class_objs.objs_df.shape[0],
                                      "N° min-max samples (g)":f"{min(n_g)} - {max(n_g)}",
                                      "N° min-max samples (r)":f"{min(n_r)} - {max(n_r)}"},
                                     ignore_index=True)
    meta_data.index = classes
    return meta_data
```

```python
check_samples = check_min_samples([RRL_objs, Ceph_objs, LPV_objs, DSCT_objs, EB_objs],
                                  ["RRL", "Ceph", "LPV", "DSCT", "EB"])
check_samples
```

4. 'QME'  -> Quadratical mutual entropy or total correlation
5. 'QMICS' -> Quadratic mutual information (QMI) based on Cauchy Schwarz distance
6. 'QMIEU' -> QMI based on Euclidean distance
